```python
import re


USE_RAZDEL = False
if USE_RAZDEL:
    from razdel import sentenize

def get_sentences(text):
    """Разбивает текст на предложения."""
    if USE_RAZDEL:
        return [sent.text.strip() for sent in sentenize(text) if sent.text.strip()]
    else:

        sentences = re.split(r'[.!?]+', text)
        return [s.strip() for s in sentences if s.strip()]

def get_words(sentence):
    """Извлекает слова (только кириллица) из предложения."""
    return re.findall(r'[а-яё]+', sentence.lower())

def average_word_length(words):
    """Средняя длина слова в символах."""
    if not words:
        return 0
    return sum(len(word) for word in words) / len(words)

def average_sentence_length_in_words(sentences):
    """Средняя длина предложения в словах."""
    if not sentences:
        return 0
    total_words = 0
    for sent in sentences:
        words = get_words(sent)
        total_words += len(words)
    return total_words / len(sentences)


pushkin_text = """
Я помню чудное мгновенье:
Передо мной явилась ты,
Как мимолетное виденье,
Как гений чистой красоты.

В томленьях грусти безнадежной,
В тревогах шумной суеты,
Звучал мне долго голос нежный
И снились милые черты.

Шли годы. Бурь порыв мятежный
Рассеял прежние мечты,
И я забыл твой голос нежный,
Твои небесные черты.

В глуши, во мраке заточенья
Тянулись тихо дни мои
Без божества, без вдохновенья,
Без слез, без жизни, без любви.

Душе настало пробужденье:
И вот опять явилась ты,
Как мимолетное виденье,
Как гений чистой красоты.

И сердце бьется в упоенье,
И для него воскресли вновь
И божество, и вдохновенье,
И жизнь, и слезы, и любовь.
"""

lermontov_text = """
Печальная Россия!
Люблю тебя — но странною любовью!
То гнев, то страх, то злоба, то раскаянье —
Всё это в душу мне твою вливает боль.

Я знаю: нет спасенья
От власти тьмы кромешной и глухой,
От гнета нищеты, от рабства произвола,
От пьянства, лжи, измен и подлости людской.

Но всё ж в тебе — надежда,
В тебе — и сила, и святая честь.
Ты не погибнешь: нет! — и в час последний
Ты воскреснешь, как феникс из пепла, честь!
"""

def analyze_text(text, author_name):
    sentences = get_sentences(text)
    all_words = []
    for sent in sentences:
        all_words.extend(get_words(sent))

    avg_word_len = average_word_length(all_words)
    avg_sent_len = average_sentence_length_in_words(sentences)

    print(f"{author_name}:")
    print(f"  Средняя длина слова: {avg_word_len:.2f} символов")
    print(f"  Средняя длина предложения: {avg_sent_len:.2f} слов")
    print()


analyze_text(pushkin_text, "Пушкин")
analyze_text(lermontov_text, "Лермонтов")
```

```python
from razdel import tokenize, sentenize
from pymorphy3 import MorphAnalyzer
from collections import Counter
import re


morph = MorphAnalyzer()

#
def get_coarse_pos(tag):
    """
    Преобразует тег pymorphy2 в обобщённую часть речи.
    Возможные значения: NOUN, VERB, ADJ, ADV, PRON, PREP, CONJ, PART, INTJ, NUM, UNKN
    """
    pos = tag.POS
    if pos == 'NOUN':
        return 'NOUN'        # существительное
    elif pos == 'VERB' or pos == 'INFN':
        return 'VERB'        # глагол (включая инфинитив)
    elif pos == 'ADJF' or pos == 'ADJS':
        return 'ADJ'         # прилагательное
    elif pos == 'ADVB':
        return 'ADV'         # наречие
    elif pos == 'PRTF' or pos == 'PRTS':
        return 'PARTICIPLE'  # причастие (можно объединить с ADJ или оставить отдельно)
    elif pos == 'GRND':
        return 'GERUND'      # деепричастие
    elif pos == 'NUMR':
        return 'NUM'         # числительное
    elif pos == 'NPRO':
        return 'PRON'        # местоимение-существительное (в pymorphy2 — NPRO)
    elif pos == 'PREP':
        return 'PREP'        # предлог
    elif pos == 'CONJ':
        return 'CONJ'        # союз
    elif pos == 'PRCL':
        return 'PART'        # частица
    elif pos == 'INTJ':
        return 'INTJ'        # междометие
    else:
        return 'UNKN'

def extract_pos_frequencies(text):
    """Извлекает частоты частей речи из текста."""

    sentences = [s.text for s in sentenize(text)]
    all_tokens = []
    for sent in sentences:
        tokens = [_.text for _ in tokenize(sent)]
        all_tokens.extend(tokens)


    words = [word.lower() for word in all_tokens if re.fullmatch(r'[а-яё]+', word.lower())]


    pos_list = []
    for word in words:
        parses = morph.parse(word)
        if parses:
            tag = parses[0].tag  # берём первый (наиболее вероятный) разбор
            coarse_pos = get_coarse_pos(tag)
            pos_list.append(coarse_pos)
        else:
            pos_list.append('UNKN')


    pos_counter = Counter(pos_list)
    total = sum(pos_counter.values())


    pos_freq = {pos: count / total for pos, count in pos_counter.items()}
    return pos_freq, total


pushkin_text = """
Я помню чудное мгновенье:
Передо мной явилась ты,
Как мимолетное виденье,
Как гений чистой красоты.

В томленьях грусти безнадежной,
В тревогах шумной суеты,
Звучал мне долго голос нежный
И снились милые черты.

Шли годы. Бурь порыв мятежный
Рассеял прежние мечты,
И я забыл твой голос нежный,
Твои небесные черты.

В глуши, во мраке заточенья
Тянулись тихо дни мои
Без божества, без вдохновенья,
Без слез, без жизни, без любви.

Душе настало пробужденье:
И вот опять явилась ты,
Как мимолетное виденье,
Как гений чистой красоты.

И сердце бьется в упоенье,
И для него воскресли вновь
И божество, и вдохновенье,
И жизнь, и слезы, и любовь.
"""

lermontov_text = """
Печальная Россия!
Люблю тебя — но странною любовью!
То гнев, то страх, то злоба, то раскаянье —
Всё это в душу мне твою вливает боль.

Я знаю: нет спасенья
От власти тьмы кромешной и глухой,
От гнета нищеты, от рабства произвола,
От пьянства, лжи, измен и подлости людской.

Но всё ж в тебе — надежда,
В тебе — и сила, и святая честь.
Ты не погибнешь: нет! — и в час последний
Ты воскреснешь, как феникс из пепла, честь!
"""

def print_pos_comparison(freq1, label1, freq2, label2, top_n=10):
    """Печатает сравнение частот частей речи."""
    all_pos = sorted(set(freq1.keys()) | set(freq2.keys()))
    print(f"{'Часть речи':<15} {label1:<10} {label2:<10} Разница")
    print("-" * 50)
    for pos in all_pos:
        f1 = freq1.get(pos, 0)
        f2 = freq2.get(pos, 0)
        diff = f2 - f1
        print(f"{pos:<15} {f1:<10.3f} {f2:<10.3f} {diff:+.3f}")


pushkin_freq, pushkin_total = extract_pos_frequencies(pushkin_text)
lermontov_freq, lermontov_total = extract_pos_frequencies(lermontov_text)

print("=== Сравнение частотности частей речи: Пушкин vs Лермонтов ===\n")
print_pos_comparison(pushkin_freq, "Пушкин", lermontov_freq, "Лермонтов")
```
